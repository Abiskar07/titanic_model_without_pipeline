```python
titanic.drop(columns=['PassengerId','Name','Ticket', 'Cabin'],inplace=True)
```

```python
titanic['family']=titanic['Parch']+titanic['SibSp']
titanic.drop(columns=['SibSp','Parch'], inplace=True)
titanic.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(titanic.loc[:,'Pclass':],titanic['Survived'],test_size=0.2, random_state=42)
```

```python
for x in x_train,x_test,y_train,y_test:
    print(x.shape)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
titanic.isnull().sum()
```

```python
imputer_age=SimpleImputer(strategy='median')
imputer_embarked=SimpleImputer(strategy='most_frequent')

x_train_age=imputer_age.fit_transform(x_train[['Age']])
x_train_embarked=imputer_embarked.fit_transform(x_train[['Embarked']])

x_test_age=imputer_age.transform(x_test[['Age']])
x_test_embarked=imputer_embarked.transform(x_test[['Embarked']])
```

```python
x_train.sample(10)
```

```python
encoder_sex=OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoder_embarked=OneHotEncoder(sparse_output=False, handle_unknown='ignore')

x_train_sex_encoded=encoder_sex.fit_transform(x_train[['Sex']])
x_train_embarked_encoded=encoder_embarked.fit_transform(x_train_embarked)

x_test_sex_encoded=encoder_sex.transform(x_test[['Sex']])
x_test_embarked_encoded=encoder_embarked.transform(x_test_embarked)

```

```python
x_train_embarked_encoded
```

```python
x_train.head()
```

```python
x_train_rem=x_train.drop(columns=['Sex','Age','Embarked'])
x_test_rem=x_test.drop(columns=['Sex','Age','Embarked'])
```

```python
x_train_embarked_encoded.shape
```

```python
import numpy as np
x_train_transformed=np.hstack((x_train_rem,x_train_age,x_train_embarked_encoded))
x_test_transformed=np.hstack((x_test_rem,x_test_age,x_test_embarked_encoded))

x_test_transformed
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree.fit(x_train_transformed,y_train)
```

```python
y_predicted=tree.predict(x_test_transformed)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predicted)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(tree,x_train_transformed, y_train,cv=10,scoring='accuracy').mean()
```
